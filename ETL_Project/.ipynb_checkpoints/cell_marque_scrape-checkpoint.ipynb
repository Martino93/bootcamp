```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
executable_path={'executable_path':'/usr/local/bin/chromedriver'}
browser=Browser('chrome',**executable_path,headless=True)
```

SCRAPING CELL MARQUE

```python
url='https://www.cellmarque.com/antibodies/primary'

cm_url = 'https://www.cellmarque.com'
browser.visit(url)
ab_names = []
img_links = []
page=1

while page<34:
    try:
        browser.click_link_by_text(str(page))
        soup=bs(browser.html,'html.parser')
        print(f"Scraping page {page}")

        for ab_index in range(8,20):    

            results = soup.find_all('div',class_='col-md-3')[ab_index]
            ab = results.find('a').text.strip().replace('\n',' ')
            ab_names.append(ab)
            #time.sleep(1)
            img = cm_url + results.find('img')['src']
            img_links.append(img)

        page+=1
        
    except IndexError as e:
        print(e)
        break
```

```python
len(img_links)
```

```python
#Cell Marque df
df = pd.DataFrame({
    'molecular_marker': ab_names,
    'img_url': img_links
})
```

```python
df['molecular_marker'].apply(lambda x: str(x).split())
```

STAINS AND MOLECULAR MARKERS BY SUBSPECIALTY

```python
#Pathology outlines
#https://www.pathologyoutlines.com/stains.html
```

```python
filepath = 'markers_subspecialty.html'
with open(filepath) as file:
    html_specialty = file.read()
    
soup_specialty=bs(html_specialty,'html.parser')
print(soup_specialty.prettify())
```

```python
print('--------- Main Categories ---------')
categories=[]
h1 = soup_specialty.find_all('h1')
for cat in range(0,len(h1)):
    categories.append(h1[cat].text)
categories
```

```python
print('--------- Sub-Categories ---------')
sub_categories=[]
h2=soup_specialty.find_all('h2')
for sub in range(0,len(h2)):
    sub_categories.append(h2[sub].text)
sub_categories
```

```python
print('--------- Bio Markers ---------')
markers=[]
a=soup_specialty.find_all('a')
for m in range(0,len(a)):
    markers.append(a[m].text)
markers
```
